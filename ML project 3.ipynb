```python
import imblearn
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
X, y = sm.fit_resample(X, y)
```

```python
from sklearn import preprocessing, tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (accuracy_score,classification_report,confusion_matrix,f1_score,mean_squared_error, precision_score, recall_score,)
from sklearn.model_selection import GridSearchCV, cross_val_score, train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVC
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, stratify=y, test_size=0.3, random_state=42)
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model_knn = KNeighborsClassifier().fit(X_train, y_train)
score_knn = model_knn.score(X_test, y_test)
y_pred_knn = model_knn.predict(X_test)
f1_score_knn = f1_score(y_test, y_pred_knn, average="macro")
precision_score_knn = precision_score(y_pred_knn, y_test, average="macro")
accuracy_score_knn = accuracy_score(y_pred_knn, y_test)
recall_score_knn = recall_score(y_pred_knn, y_test, average="macro")

cm_knn = confusion_matrix(y_test, y_pred_knn)
sns.heatmap(cm_knn, annot=True, fmt=".0f",linewidths=3,square=True, cmap="Greens", color="#cd1076",)
plt.ylabel("actual label", color="blue")
plt.xlabel("predicted label", color="blue")
plt.title(f"F1: {f1_score_knn:.2f}", size=14, color="red")
plt.show()
```

```python
model_knn.score(X_train, y_train), model_knn.score(X_test, y_test)
```

```python
model_svc = SVC().fit(X_train, y_train)
score_svc = model_svc.score(X_test, y_test)
y_pred_svc = model_svc.predict(X_test)


f1_score_svc = f1_score(y_test, y_pred_svc, average="macro")
precision_score_svc = precision_score(y_pred_svc, y_test, average="macro")
accuracy_score_svc = accuracy_score(y_pred_svc, y_test)
recall_score_svc = recall_score(y_pred_svc, y_test, average="macro")
cm_svc = confusion_matrix(y_test, y_pred_svc)
sns.heatmap(cm_svc,annot=True,fmt=".0f", linewidths=3,square=True, cmap="Greens",color="#cd1076",)
plt.ylabel("actual label", color="blue")
plt.xlabel("predicted label", color="blue")
plt.title(f"F1: {f1_score_svc:.2f}", size=14, color="red")
plt.show()
```
